```python
import os
import glob
import pandas as pd
import numpy as np

youtube = pd.read_csv('E:/DataScience/youtube.csv')
newyoutube = youtube.groupby(['Track Name', 'Artist Names', 'YouTube URL'])['Views'].sum().reset_index()
newyoutube1 = newyoutube.head(120)
newyoutube1
```

```python
from bs4 import BeautifulSoup
import requests
def scrape_info(url):
      
    # getting the request from url
    r = requests.get(url)
      
    # converting the text
    s = BeautifulSoup(r.text, "html.parser")
      
    # finding meta info for views
    views = s.find("div", class_="view-count style-scope ytd-video-view-count-renderer")
      
    # returning the dictionary
    return views
```
