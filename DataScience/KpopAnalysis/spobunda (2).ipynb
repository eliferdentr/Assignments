```python
def getArtistData (id):
    
    top10Songs_list = ["Artist Name","Track Name", "Track Number","Popularity", "Song ID","danceability","energy","key","loudness","mode", "speechiness","instrumentalness","liveness","valence","tempo", "duration_ms","time_signature"]
    top10Songs_df = pd.DataFrame(columns = top10Songs_list)
    
    results = spotify.artist_top_tracks(id)
    

    for track in results['tracks'][:10]:
        top10Songs1 = {}
        top10Songs1["Track Name"] = track['name']
        top10Songs1["Song ID"] = track['id']
        top10Songs1["Popularity"] = track['popularity']
        top10Songs1["Track Number"] = track['track_number']
        top10Songs1["Artist Name"] = track['artists'][0]['name']
        
        audio_features = spotify.audio_features(top10Songs1["Song ID"])[0]
        for feature in top10Songs_list[5:]:
            top10Songs1[feature] = audio_features[feature]
  
        # Concat the dfs
        track_df = pd.DataFrame(top10Songs1, index = [0])
        top10Songs_df = pd.concat([top10Songs_df, track_df], ignore_index = True)
        
    return top10Songs_df 
```

```python
EXO = 'spotify:artist:3cjEqqelV9zb4BYE3qDQ4O'
EXO_df = getArtistData (EXO)

EXO_df.sort_values(["Popularity"], ascending=False)
EXO_df
```
