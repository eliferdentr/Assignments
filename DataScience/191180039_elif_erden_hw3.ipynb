```python
from sklearn import datasets
import pandas as pd


iris = datasets.load_iris()

X = pd.DataFrame(iris.data)
y = pd.DataFrame(iris.target)

```

```python
#splitting the data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, shuffle = True, random_state = 123)
```

```python

import graphviz
```

```python
#Training a Decision-Tree Classifier with Gini index -1
from sklearn.tree import DecisionTreeClassifier
from sklearn import tree
decision_tree_model = DecisionTreeClassifier()
decision_tree_model = decision_tree_model.fit(X_train,y_train)
#A2: When calculating the amount of probability of a specific feature that is classified incorrectly when selected randomly, gini index is used.
#l-2: https://medium.com/analytics-steps/understanding-the-gini-index-and-information-gain-in-decision-trees-ab4720518ba8
```

```python

dot_data = tree.export_graphviz(decision_tree_model, out_file=None, feature_names=["petal_length", "petal_width", "sepal_length"," sepal_width"], class_names=["setosa", "versicolor", "virginica"], filled=True, rounded=True, special_characters=True)
graph = graphviz.Source(dot_data)
graph
```

```python
 #predicting the species of each flower in the test set -1
pred = decision_tree_model.predict(X_test)
pred
```

```python
#confusion matrix for the predictions -1
from sklearn import metrics
metrics.confusion_matrix(pred, y_test)
#A4: Confusion matrix is used for visualization of the performance of an algorithm
#l-4: https://en.wikipedia.org/wiki/Confusion_matrix
```

```python
#accuracy -1
from sklearn.metrics import accuracy_score
acc= accuracy_score(pred, y_test)
print('Accuracy: %.3f' % acc)
#A5: By using accuracy, the best model for identifying relationships and patterns between variables in a dataset can be chosen.
#l-5:https://www.datarobot.com/wiki/accuracy/
#The model is overfitting because the accuracy measured against the training set is high
```

```python
#Precision -1
from sklearn.metrics import precision_score
precision = precision_score(pred, y_test, average='micro')
print('Precision: %.3f' % precision)
#A6:Precision is the  ratio between the True Positives and (True Positives + False Positives)
#l-6:https://developers.google.com/machine-learning/crash-course/classification/precision-and-recall
```

```python
#Recall -1
from sklearn.metrics import recall_score
recall = recall_score(pred, y_test, labels=[1,2], average='micro')
print('Recall: %.3f' % recall)
#A7: Recall is the  ratio between the True Positives and (True Positives + False Negatives)
#l-7:https://developers.google.com/machine-learning/crash-course/classification/precision-and-recall
```

```python
#f1 score -1
import numpy as np
from sklearn.metrics import f1_score
f1 = f1_score(y_test, pred, average='micro')
print('f1 score: %.3f' % f1)
#A-8: f1 score is the measure of a model's accuracy on a dataset.
#l-8: https://www.datarobot.com/wiki/accuracy/
```

```python
#Training a a Decision-Tree Classifier with entropy criterion. -2
decision_tree_model1 = DecisionTreeClassifier(criterion = "entropy")
decision_tree_model1 = decision_tree_model.fit(X_train,y_train)

#A1: Information gain calculates the entropy reduction. It is is used for determining the best features that shows the most information about a class.
#A3: Entropy is either the measure of the purity or degree of uncertainty of a random variable.
#link 1-3 : https://www.analyticssteps.com/blogs/what-gini-index-and-information-gain-decision-trees
```

```python
dot_data1 = tree.export_graphviz(decision_tree_model1, out_file=None, feature_names=["petal_length", "petal_width", "sepal_length"," sepal_width"], class_names=["setosa", "versicolor", "virginica"], filled=True, rounded=True, special_characters=True)
graph1 = graphviz.Source(dot_data1)
graph1
```

```python
#predicting the species of each flower in the test set. -2
pred1 = decision_tree_model1.predict(X_test) #Predictions on Testing data
pred1
```

```python
#confusion matrix for the predictions -1
metrics.confusion_matrix(pred1, y_test)
```

```python
#Accuracy -2
acc1= accuracy_score(pred1, y_test)
print('Accuracy: %.3f' % acc1)
```

```python
#Precision -2
precision1 = precision_score(pred1, y_test, average='micro')
print('Precision: %.3f' % precision1)
```

```python
#Recall -2
recall1 = recall_score(pred1, y_test, labels=[1,2], average='micro')
print('Recall: %.3f' % recall1)
```
