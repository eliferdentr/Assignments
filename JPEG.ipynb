```python
classes = ["person"]
```

```python
def calculate_efficiency(image_path):
    original_image = Image.open(image_path)
    original_size = os.path.getsize(image_path)
    compressed_image_dir = "/content/drive/MyDrive/openimages/compressed"
    filename = os.path.splitext(os.path.basename(image_path))[0]
    compressed_image_path = os.path.join(compressed_image_dir, filename + ".jpeg")

    start_time = time.time()
    original_image.save(compressed_image_path, format="JPEG", quality=90)
    end_time = time.time()
    compressed_size = os.path.getsize(compressed_image_path)


    compression_ratio = original_size / compressed_size

    computational_complexity = end_time - start_time

    return compression_ratio, computational_complexity
```

```python
import time
import os
from PIL import Image
```

```python
cr_sum = 0.0
cc_sum = 0.0
total_images = 0
save_dir = "/content/drive/MyDrive/openimages"
```

```python
for class_name in classes:
    class_dir = os.path.join(save_dir, class_name, 'images')
    for filename in os.listdir(class_dir):
        if filename.endswith(".jpg"):
            total_images += 1
            image_path = os.path.join(class_dir, filename)
            compression_ratio, computational_complexity = calculate_efficiency(image_path)
            cr_sum += compression_ratio
            cc_sum += computational_complexity
            print(f"Image: {filename}")
            print(f"Class: {class_name}")
            print(f"Compression Ratio: {compression_ratio}")
            print(f"Computational Complexity: {computational_complexity} seconds")
            print("-----------------------------")

```

```python
cr_percentage = (cr_sum / total_images) * 100
cc_percentage = (cc_sum / total_images) * 100

print(f"Total Images: {total_images}")
print(f"Sum of Compression Ratio: {cr_sum}")
print(f"Sum of Computational Complexity: {cc_sum} seconds")
print(f"Compression Ratio Percentage: {cr_percentage}%")
print(f"Computational Complexity Percentage: {cc_percentage}%")
```
